```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named df
# ... (your DataFrame creation code)

df_diab = df[df['Diabetes_binary'] == 1.0]
# Set the color palette to green
colors = ["#1c451f"]
  # Set your custom color palette
sns.set_palette(sns.color_palette(colors))
# Specify the columns
col = ['Age', 'Sex', 'HighChol', 'CholCheck', 'Smoker',
       'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk',
       'Stroke', 'HighBP', 'BMI']

for i in col:
    print('------------------------------------------')
    print(i)



    if i in ['BMI', 'MentHlth', 'PhysHlth']:
        # Plot histogram for continuous variables
        plt.hist(df_diab[i], bins=10, label=i)
    else:
        # Plot bar plot for categorical variables
        sns.countplot(x=i, data=df_diab,color='#1c451f')

    plt.xlabel(i)
    plt.ylabel('Count')
    plt.title(f'{i} Distribution for Diabetes_binary=1.0')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df
# ... (your DataFrame creation code)

x = 'Diabetes_binary'

# Group by 'Diabetes_binary' and calculate the percentage distribution
df1 = df[x].value_counts(normalize=True).mul(100).reset_index(name='percent')

print(f"{x} vs percent")

# Set up the figure and axes
fig, ax = plt.subplots()

# Use seaborn barplot to visualize the percentages
sns.barplot(x='index', y='percent', data=df1, ax=ax,color='#1c451f')

# Annotate bars with percentages
for p in ax.patches:
    txt = f"{p.get_height().round(2)}%"
    txt_x = p.get_x() + p.get_width() / 2
    txt_y = p.get_height()
    ax.text(txt_x, txt_y, txt, ha='center')

plt.ylim(0, 100)
plt.xlabel(x)
plt.ylabel('Percentage')
plt.title(f"{x} vs percent")
plt.show()
```

**Bivariate Analysis**

```python
for x in ['Age', 'Sex', 'HighChol', 'CholCheck', 'Smoker',
       'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk',
       'Stroke', 'HighBP']:
       y='Diabetes_binary'
       df1 = df.groupby(y)[x].value_counts(normalize=True)
       df1 = df1.mul(100)
       df1 = df1.rename('percent').reset_index()
       print(x+" vs percent")
       colors = ["#1c451f", "#5eb565"]
      # Set your custom color palette
       sns.set_palette(sns.color_palette(colors))


       g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df1)
       g.ax.set_ylim(0,100)
       plt.show()
       for p in g.ax.patches:
        txt = str(p.get_height().round(2)) + '%'
        txt_x = p.get_x()
        txt_y = p.get_height()
        g.ax.text(txt_x,txt_y,txt)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame

for x in ['MentHlth', 'PhysHlth','Age','Income']:
    y = 'Diabetes_binary'
    df1 = df.groupby(y)[x].value_counts(normalize=True).mul(100).rename('percent').reset_index()
    colors = ["#1c451f", "#5eb565"]
      # Set your custom color palette
    sns.set_palette(sns.color_palette(colors))

    plt.figure(figsize=(20, 10))
    sns.barplot(x=x, y='percent', hue=y, data=df1)

    plt.title(f'{x} vs percent')
    plt.ylim(0, 70)
    plt.show()
```

```python


# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(25, 13))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

      # Set your custom color palette
sns.heatmap(correlation_matrix, annot=True, cmap='Greens',mask=mask)
plt.title('Correlation Heatmap')
plt.show()
```
